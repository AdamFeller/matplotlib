```python
import matplotlib.pyplot as plt
import csv
```

### Example 1 : a bit messy

```python
x = []
y = []

with open ('example.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
        

plt.plot(x,y, label = 'Loaded from file!')

plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting Graph \n Check it out')
plt.legend()
plt.show()
```

### Example 2 :

```python
import numpy as np
```

```python
# unpack = True means that I want to unpack the elements seperated by that delimiter into x and y
# so you will need the same number of variables as elements to unpack
x,y = np.loadtxt('example.txt', delimiter = ',', unpack = True)

plt.plot(x,y, label = 'Loaded from file!')

plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting Graph \n Check it out')
plt.legend()
plt.show()
```
