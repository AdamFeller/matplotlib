```python
import matplotlib.pyplot as plt
```

```python
slices = [7,2,2,13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c','m','r','b']

plt.pie(slices, 
        labels = activities, 
        colors = cols, 
        startangle =90, 
        shadow = True, 
        explode = (0,0,0.1,0),
       autopct = '%1.1f%%')

#plt.xlabel('Plot Number')
#plt.ylabel('Important var')
plt.title('Interesting Graph \n Check it out')
#plt.legend()
plt.show()
```
