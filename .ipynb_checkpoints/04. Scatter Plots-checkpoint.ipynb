```python
import random 

x = [1,2,3,4,5,6,7,8]
y = [random.randint(0,100) for x in range(len(x))]
y2 = [random.randint(0,100) for x in range(len(x))]
y3 = [random.randint(0,100) for x in range(len(x))]


plt.scatter(x,y, label = 'Skitscat', color ='k', marker = '$hello!$', s= 1000)
plt.scatter(x,y2, label = 'Skitscat', color ='b', marker = '*', s= 100)
plt.scatter(x,y3, label = 'Skitscat', color ='g', marker = '_', s= 400)

plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting Graph \n Check it out')
plt.legend()
plt.show()
```

```python
# Note in the title the '\n' is for a new line
```
